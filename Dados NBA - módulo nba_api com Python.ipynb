Importando as bibliotecas necessárias:

```python
import pandas as pd
import numpy as np

from nba_api.stats.static import players
from nba_api.stats.static import teams 

import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
from matplotlib import pyplot
```

Algumas tabelas são grandes, então vamos configurar o notebook para expandir os dados das colunas:

```python
pd.set_option('display.max_columns',250)
```

## Obtendo dados dos times

Podemos criar um dicionário com o nome dos times e sua identificação numérica. A idenficação pode ser usada para acessar dados do time de interesse.

```python
# use o get_teams para retornar uma lista de 30 dicionários, cada um com um time da NBA.
teams_dict = teams.get_teams()
```

```python
#podemos organizar estes dados em um dataframe
teams = pd.DataFrame(teams_dict)
teams
```

## Obtendo dados dos jogadores

Da mesma forma que o item anterior, vamos criar um novo dicionário, que vai conter os dados de todos os jogadores que já jogaram na NBA.

```python
# use o get_players para retornar uma lista com todos os jogadores da NBA (em atividade ou não).
player_dict = players.get_players()
```

```python
# vamos organizar estes dados em um dataframe
players = pd.DataFrame(player_dict)
players
```

## Investigando os jogadores

Com os dados de todos os jogadores em mãos, podemos procurar os jogadores de interesse.

```python
# buscando dados de um jogador usando o primeiro nome
james = players[players['first_name'] == 'LeBron'] 
james
```

```python
# buscando dados de um jogador usando o último nome
giannis = players[players['last_name'] == 'Antetokounmpo'] 
giannis
```

```python
# buscando dados de um jogador usando o nome completo
kobe = players[players['full_name'] == 'Kobe Bryant']
kobe
```

Podemos também fazer um dataframe com todos os jogadores ativos no momento:

```python
active = players[players['is_active'] == True]
active
```

## Dados da carreira de cada jogador

Para investigar a carreira de um jogador de interesse, vamos importar o PlayerCareerStats 

https://github.com/swar/nba_api/blob/master/docs/nba_api/stats/endpoints/playercareerstats.md

```python
from nba_api.stats.endpoints import playercareerstats
```

Explorando a carreira do Kobe Bryant:

```python
# usaremos o número de identificação do Kobe Bryant
career_kobe = playercareerstats.PlayerCareerStats(player_id='977')
kobe_all = career_kobe.get_data_frames()[0]
kobe_all
```

Para entender as abreviações: https://en.wikipedia.org/wiki/Basketball_statistics

Agora que temos o dataframe, podemos visualizar os dados:

```python
k = sns.catplot(data=kobe_all, x='SEASON_ID', y='PTS', kind='bar', height=8.27, aspect=12.7/8.27)
plt.xlabel("Temporada")
plt.ylabel("Pontos")
plt.title("Kobe Bryant: pontos em cada temporada")
plt.show(k)
```

- Kobe Bryant não participou de toda a temporada de 2013-14 por uma lesão no joelho. 

## Dados sobre os jogos por jogador

Podemos usar a identificação do jogador obtida nas tabelas anteriores e criar um dataframe com todos os jogos de um determinado jogador, usando o PlayerGameLog

https://github.com/swar/nba_api/blob/master/docs/nba_api/stats/endpoints/playergamelog.md

```python
from nba_api.stats.endpoints import playergamelog
```

```python
#obtendo a identificação de cada jogo usando o id do jogador Kobe Bryant
gamelog_player = playergamelog.PlayerGameLog(player_id='977', season='ALL')
```

```python
# transforma os dados em um dataframe
kobe_games = gamelog_player.get_data_frames()[0]
kobe_games
```

Podemos separar as 15 maiores pontuações por jogo com o código abaixo:

```python
# usando sort_values para ordenar os pontos e obtendo as 15 maiores pontuações
kobe_games1 = kobe_games.sort_values(by='PTS', ascending=False).iloc[:15]
kobe_games1
```

Usamos o dataframe com as 15 maiores pontuações para vizualização:

```python
k1 = sns.catplot(data=kobe_games1, x='PTS', y='Game_ID', kind='bar', orient="h", height=8.27, aspect=12.7/8.27)
plt.xlabel("Pontos")
plt.ylabel("Identificação dos jogos")
plt.title("Kobe Bryant: 15 maiores pontuações por jogo")
plt.show(k1)
```

## Dados de todas as jogadas por jogo

A API também disponibiliza o PlayByPlay, que é a descrição de todas as jogadas por jogo.

https://github.com/swar/nba_api/blob/master/docs/nba_api/stats/endpoints/playbyplay.md

```python
from nba_api.stats.endpoints import playbyplay
```

Para este exemplo, vou usar o jogo com o maior número de pontos do Kobe Bryant, que foi definido no dataframe anterior.

```python
playbyplay = playbyplay.PlayByPlay(game_id='0020500591').get_data_frames()[0]
playbyplay
```

## Dados de cada jogada por jogador e por temporada

Para encontrar os dados de cada jogada que um determinado jogador fez em uma temporada, vamos usar o PlayerGameLogs (que é diferente do PlayerGameLog que usamos acima):

https://github.com/swar/nba_api/blob/master/docs/nba_api/stats/endpoints/playergamelogs.md

```python
from nba_api.stats.endpoints import playergamelogs
```

```python
gamelogs_player = playergamelogs.PlayerGameLogs(player_id_nullable='977', season_nullable='2005-06')
kobe_gamelogs = gamelogs_player.get_data_frames()[0]
kobe_gamelogs
```
